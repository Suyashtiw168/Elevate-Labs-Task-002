```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the Titanic dataset

df = pd.read_csv('Titanic Dataset Intern T 02.zip')

# looking at the data
print("üîç First 5 rows of the dataset:")
print(df.head())

# Basic statistics
print("\nüìä Summary statistics:")
print(df.describe())

# Checking for missing values
print("\n‚ùì Missing values:")
print(df.isnull().sum())

# Ploting histograms for numeric columns
df.hist(figsize=(10, 8))
plt.suptitle("üìâ Histograms of Numeric Features", fontsize=16)
plt.tight_layout()
plt.show()

# Boxplots to find outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title("üì¶ Boxplots of Age and Fare")
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("üìà Correlation Heatmap")
plt.show()

# 7. Pairplot to explore feature relationships 
# sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']].dropna(), hue='Survived')

# 8. Some quick visual insights
sns.countplot(x='Survived', data=df)
plt.title("üßç Survival Count")
plt.show()

sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("üö¢ Survival by Class")
plt.show()
```
